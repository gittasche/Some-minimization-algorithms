```python
import numpy as np
from unconstrained.conjugate_gradients import ConjugateGradients
```

```python
CG_FR = ConjugateGradients(method = 'Fletcher-Reeves')
Ros_func = lambda x: 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2
init = np.array([-4.1, -4.])
fin = CG_FR.optimize(Ros_func, init)
with np.printoptions(precision=5, suppress=True):
    print(f'Minima: [x, y] = {fin}')
print(f'Iteration count: {CG_FR.num_steps_}')
CG_FR.plot(Ros_func, CG_FR.points_, x_min=-5, x_max=5, y_min=-5, y_max=5)
```

```python
CG_PR = ConjugateGradients(method = 'Polak-Ribiere')
Ros_func = lambda x: 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2
init = np.array([-4.1, -4.])
fin = CG_PR.optimize(Ros_func, init)
with np.printoptions(precision=5, suppress=True):
    print(f'Minima: [x, y] = {fin}')
print(f'Iteration count: {CG_PR.num_steps_}')
CG_PR.plot(Ros_func, CG_PR.points_, x_min=-5, x_max=5, y_min=-5, y_max=5)
```
