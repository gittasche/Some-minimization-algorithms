```python
import numpy as np
from unconstrained.l_bfgs import L_BFGS
```

```python
LBFGS = L_BFGS()
Ros_func = lambda x, y: 100*(y - x**2)**2 + (1 - x)**2
init = np.array([-4.1, -4.])
fin = LBFGS.optimize(Ros_func, init)
with np.printoptions(precision=5, suppress=True):
    print(f'Minima: [x, y] = {fin}')
print(f'Iteration count: {LBFGS.num_steps_}')
LBFGS.plot(Ros_func, LBFGS.points_, x_min=-5, x_max=5, y_min=-5, y_max=5)
```
