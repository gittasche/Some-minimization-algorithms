```python
import numpy as np
from unconstrained.bfgs import BFGS
```

```python
bfgs = BFGS()
Ros_func = lambda x: 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2
init = np.array([-4.1, -4.])
fin = bfgs.optimize(Ros_func, init)
with np.printoptions(precision=5, suppress=True):
    print(f'Minima: [x, y] = {fin}')
print(f'Iteration count: {bfgs.num_steps_}')
bfgs.plot(Ros_func, bfgs.points_, x_min=-5, x_max=5, y_min=-5, y_max=5)
```
